```python
def submarine_move_part1(file):
    depth = 0
    dist = 0
    for index, row in enumerate(file):
        if 'forward' in row:
            dist += int(row[len('forward')+1:])
        elif 'up' in row:
            depth -= int(row[len('up')+1:])
        elif 'down' in row:
            depth += int(row[len('down')+1:])
        else:
            print('unknown command')
    print(str(depth*dist))

f = open("submarine_movements.txt", "r")
submarine_move_part1(f)
f.close()
```
