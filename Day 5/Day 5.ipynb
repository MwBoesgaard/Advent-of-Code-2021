```python
import numpy as np
import matplotlib.pyplot as plt

def hydrothermal_scanner(file, size_of_map):
    file_content = file.read().split('\n')
    
    map_of_vents = create_map(size_of_map**2)
    
    for row in file_content:
        parsed_row = row_parser(row)
        if instruction_is_horizontal_or_vertical(parsed_row):
            points_to_map = coordinates_to_update(parsed_row)
            map_points_to_map(map_of_vents, points_to_map)
            
    print_map(map_of_vents)
    ans = num_of_intersections(map_of_vents)
    print(f"Number of intersections: {ans}")
    
def create_map(size):
    sqrt_size = int(np.sqrt(size))
    array = np.zeros((sqrt_size, sqrt_size))
    return array

def row_parser(row):
    row_components = row.split('->')
    row_components[0].strip()
    row_components[1].strip()
    
    start_cord = row_components[0].split(',')
    end_cord = row_components[1].split(',')
    
    start_cord = [start_cord[0].strip(), start_cord[1].strip()]
    end_cord = [end_cord[0].strip(), end_cord[1].strip()]
    
    return [start_cord, end_cord]
    
def instruction_is_horizontal_or_vertical(parsed_row):
    if parsed_row[0][0] == parsed_row[1][0] or parsed_row[0][1] == parsed_row[1][1]:
        return True
    else:
        return False
    
def print_map(map_of_vents):
    fig, ax = plt.subplots(figsize = (14,14))
    ax.imshow(map_of_vents, interpolation='none')
    plt.show()

def coordinates_to_update(parsed_row):
    x1 = int(parsed_row[0][0])
    y1 = int(parsed_row[0][1])
    x2 = int(parsed_row[1][0])
    y2 = int(parsed_row[1][1])
    
    change_x = x2 - x1
    change_y = y2 - y1
    
    list_of_coordinates = []
    
    step_x = 1
    step_y = 1
    
    if change_x < 0:
        step_x = -1
        
    if change_y < 0:
        step_y = -1

    
    if (change_x != 0):
        for delta in range(0, change_x+step_x, step_x):
            list_of_coordinates.append([x1+delta, y1])
    elif (change_y != 0):
        for delta in range(0, change_y+step_y, step_y):
            list_of_coordinates.append([x1, y1+delta])
    return list_of_coordinates

def map_points_to_map(map_of_vents, coordinates):
    for cords in coordinates:
        x_cord = cords[0]
        y_cord = cords[1]
        map_of_vents[y_cord][x_cord] += 1
    return map_of_vents

def num_of_intersections(map_of_vents):
    counter = 0
    for row in map_of_vents:
        for col in row:
            if col > 1:
                counter += 1
    return counter

file = open('hydrothermal_data.txt', 'r')
hydrothermal_scanner(file, 1000)
file.close()
```
