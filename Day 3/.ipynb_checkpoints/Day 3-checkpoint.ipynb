```python
def diagnostic_rutine_part1(file):
    length_of_file = 0
    dict_of_bits = {}

    for index, row in enumerate(file):
        row_list = list(row)
        row_list.pop() #to remove a newline ('\n')
        list_len = len(row_list)

        for i, bit in enumerate(row_list):
            if str(i) in list(dict_of_bits.keys()):
                dict_of_bits[str(i)] += int(bit)
            else:
                dict_of_bits[str(i)] = int(bit)
        
        length_of_file = index
        
    length_of_file += 1
    print(length_of_file, dict_of_bits)

    gamma_list = [int(round(x/length_of_file)) for x in dict_of_bits.values()]
    epsilon_list = [1-x for x in gamma_list]
    
    gamma_list = [str(x) for x in gamma_list]
    epsilon_list = [str(x) for x in epsilon_list]

    gamma = ''.join(gamma_list)
    epsilon = ''.join(epsilon_list)
    
    print("Gamma: " + gamma + ", Epsilon: " + epsilon)
    
    gamma = int(gamma, 2)
    epsilon = int(epsilon, 2)
    
    print("Gamma: " + str(gamma) + ", Epsilon: " + str(epsilon))
    
    #Power consumption
    return (gamma * epsilon)


f = open("diagnostic_data.txt", "r")
print("Power consumption of the submarine is: " + str(diagnostic_rutine_part1(f)))
f.close()

```

```python
def diagnostic_rutine_part2(file_contents):
    O2 = O2_CO2_meter(file_contents, 0, "O2")
    CO2 = O2_CO2_meter(file_contents, 0, "CO2")
    print("O2: " + str(O2), "CO2: " + str(CO2))
    return O2 * CO2
        
def O2_CO2_meter(file_contents, pos, target):
    
    print(len(file_contents), pos, target)

    if len(file_contents) == 1:
        ans = file_contents[0]
        print(ans.replace("\n", ""))
        ans = int(ans, 2)
        print(target + ": " + str(ans))
        return ans
    
    length_of_file, dict_of_bits = dict_maker(file_contents)

    if target == "O2":
        if dict_of_bits[str(pos)]/len(file_contents) == 0.5:
            common_bit = 1
        else:
            common_bit = int(round(dict_of_bits[str(pos)]/length_of_file))
    elif target == "CO2":
        if dict_of_bits[str(pos)]/len(file_contents) == 0.5:
            common_bit = 0
        else:
            common_bit = 1 - int(round(dict_of_bits[str(pos)]/length_of_file))
    else:
        raise Exception('Wrong target value!')

    new_file_contents = input_filter(file_contents, pos, str(common_bit))
    return O2_CO2_meter(new_file_contents, pos+1, target)

def dict_maker(file_contents):
    length_of_file = 0
    dict_of_bits = {}
    
    for index, row in enumerate(file_contents):
        row_list = list(row)
        row_list.pop() #to remove a newline ('\n')
        list_len = len(row_list)

        for i, bit in enumerate(row_list):
            if str(i) in list(dict_of_bits.keys()):
                dict_of_bits[str(i)] += int(bit)
            else:
                dict_of_bits[str(i)] = int(bit)
        
        length_of_file = index
        
    
    length_of_file += 1
    
    return length_of_file, dict_of_bits

def input_filter(file_contents, pos, cond):
    return [row for row in file_contents if row[pos] == cond]
    
f = open("diagnostic_data.txt", "r")
file_contents = f.readlines()
print("O2 generation and CO2 scubbing: " + str(diagnostic_rutine_part2(file_contents)))
f.close()

```
