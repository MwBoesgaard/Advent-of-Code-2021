```python
def sonar_sweep_part1(file):
    file_content = file.read().split('\n')
    previous = 0
    count = 0
    for index, x in enumerate(file_content):
        x = int(x)
        status = "N/A - no previous measurement"
        if (index > 0):
            if (x > previous):
                status = "increased"
                count += 1
            if (x == previous):
                status = "no change"
            if (x < previous):
                status = "decreased"
        previous = x
        print(f"{x}, {status}")
    print("*****************************")
    print(f"The number of increased measurements is {count}")
    
    
f = open("sonar_data.txt", "r")
sonar_sweep_part1(f)
f.close()
```

```python
def sonar_sweep_part2(file):
    file_content = file.read().split('\n')
    previous = 0
    count = 0
    three_recent_measurements = []
    for index, x in enumerate(file_content):
        three_recent_measurements.append(int(x))
        sum_of_mes = sum(three_recent_measurements)
        len_of_mes = len(three_recent_measurements)
        avg = int(sum_of_mes/len_of_mes)
        status = "N/A - no previous sum of 3"
        if (index > 2):
            if (sum_of_mes > previous):
                status = "increased"
                count += 1
            if (sum_of_mes == previous):
                status = "no change"
            if (sum_of_mes < previous):
                status = "decreased"
        previous = sum_of_mes
        if len(three_recent_measurements) == 3:
            three_recent_measurements = three_recent_measurements[1:]
        print(f"{avg}, {status}")
    print("*****************************")
    print(f"Result: {count}")
    
f = open("sonar_data.txt", "r")
sonar_sweep_part2(f)
f.close()
```
